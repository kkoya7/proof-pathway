```python
# Cell 1 вЂ” lazy + pushdown
import polars as pl, time
t0 = time.perf_counter()
out = (
    pl.scan_parquet("/Users/kaustubhkoya/Policy Research/Datasets/medicaid-provider-spending.parquet")
    .filter(pl.col("HCPCS_CODE") == "83036")
    .group_by("CLAIM_FROM_MONTH")
    .agg(pl.sum("TOTAL_PATIENTS").alias("patients"))
    .collect()
)
print(f"lazy: {time.perf_counter()-t0:.2f}s, shape {out.shape}")
```
